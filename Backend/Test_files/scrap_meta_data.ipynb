## Get meta-data


```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.3'
}
YEAR_LIST = [2014, 2015, 2016, 2017, 2018, 2019]

# get list of url from maxim's github
for YEAR in YEAR_LIST:
    print(f'* ____________________________ STARTING YEAR: {YEAR} ____________________________ *')
    url = f"https://raw.githubusercontent.com/GraphtyLove/KPMG-Challenge/master/assets/json/links_entreprises/links_entreprises_{YEAR}.json"
    response = urlopen(url)
    data = json.loads(response.read())
    data = list(set(data))
    meta_data = {}
    for i, link in enumerate(data[:5]):
        try:
            time.sleep(0.1)
            reg_url = link
            req = Request(url=reg_url, headers=headers) 
            html = urlopen(req).read()
            soup = BeautifulSoup(html,'lxml')

            title = soup.find_all('td',attrs={"class":"title"}, text=True)
            data = soup.find_all('td', attrs={"colspan": "2"}, text=True)

            for k,v in enumerate(title):
                value = data[k].getText()
                value_as_string = str(value)
                if "Laatste" not in value_as_string:
                    meta_data[v.getText()] = value_as_string

            # !!!! ADD PDF NUMBER HERE !!!!
            with open(f'../assets/json/meta-data/{YEAR}/PDF-NUMBER-HERE.meta.json', 'w', encoding='utf-8') as file:
                json.dump(meta_data, file, ensure_ascii=False, indent=4)

        except:
            print(f'ERROR url: {i} {link}')
            print(sys.exc_info())
```
